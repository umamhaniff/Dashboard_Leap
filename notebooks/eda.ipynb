# Exploratory Data Analysis (EDA) Notebook

This notebook performs exploratory data analysis on the Dashboard Leap dataset.

```python
# Import required libraries
import sys
import os
sys.path.append(os.path.join(os.getcwd(), 'src'))

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from src.data.loader import load_data
from src.data.cleaner import clean_data
from src.data.transformer import transform_data
from src.visualization.plot import create_histogram, create_correlation_heatmap

# Set plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

```python
# Load data
print("Loading data...")
data = load_data()
print(f"Data shape: {data.shape}")
print(f"Columns: {list(data.columns)}")
data.head()
```

```python
# Data cleaning
print("Cleaning data...")
cleaned_data = clean_data(data)
print(f"Cleaned data shape: {cleaned_data.shape}")
print(f"Missing values after cleaning: {cleaned_data.isnull().sum().sum()}")
```

```python
# Basic statistics
print("Basic statistics:")
cleaned_data.describe()
```

```python
# Data types and missing values
print("Data types and missing values:")
data_info = pd.DataFrame({
    'Data Type': cleaned_data.dtypes,
    'Non-Null Count': cleaned_data.count(),
    'Null Count': cleaned_data.isnull().sum(),
    'Null Percentage': (cleaned_data.isnull().sum() / len(cleaned_data)) * 100
})
data_info
```

```python
# Correlation analysis
numeric_cols = cleaned_data.select_dtypes(include=[np.number]).columns
if len(numeric_cols) > 1:
    correlation_matrix = cleaned_data[numeric_cols].corr()
    print("Correlation matrix:")
    correlation_matrix
```

```python
# Visualize correlations
if len(numeric_cols) > 1:
    fig = create_correlation_heatmap(cleaned_data)
    plt.show()
```

```python
# Distribution analysis
for col in numeric_cols[:5]:  # Analyze first 5 numeric columns
    print(f"\nAnalyzing {col}:")
    print(f"Mean: {cleaned_data[col].mean():.2f}")
    print(f"Median: {cleaned_data[col].median():.2f}")
    print(f"Std: {cleaned_data[col].std():.2f}")
    print(f"Min: {cleaned_data[col].min():.2f}")
    print(f"Max: {cleaned_data[col].max():.2f}")
    
    # Create histogram
    fig = create_histogram(cleaned_data, col)
    plt.show()
```

```python
# Categorical analysis
categorical_cols = cleaned_data.select_dtypes(include=['object', 'category']).columns
for col in categorical_cols[:3]:  # Analyze first 3 categorical columns
    print(f"\nAnalyzing {col}:")
    print(f"Unique values: {cleaned_data[col].nunique()}")
    print("Value counts:")
    print(cleaned_data[col].value_counts().head(10))
```

```python
# Data transformation
print("Transforming data...")
transformed_data = transform_data(cleaned_data, create_features=True)
print(f"Transformed data shape: {transformed_data.shape}")
print(f"New columns added: {set(transformed_data.columns) - set(cleaned_data.columns)}")
```

```python
# Save processed data
output_dir = "data/processed"
os.makedirs(output_dir, exist_ok=True)

cleaned_data.to_csv(os.path.join(output_dir, "cleaned_data.csv"), index=False)
transformed_data.to_csv(os.path.join(output_dir, "transformed_data.csv"), index=False)

print("Processed data saved to data/processed/")
```
